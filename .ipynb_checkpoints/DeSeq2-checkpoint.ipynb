#### MA plot
An MA plot show the mean expression vs the log fold change and is a very common method to visualize the results of a differential expression plot. Significantly differentially expressed genes are colored red:

```r
options(repr.plot.width=14, repr.plot.height=7)
plotMA(dds,ylim=c(-3,3))
#this one has to be in your notebook for the project


#the x axis is the l(og(stim)+log(unstim))/2
#the y axis is the stimulated over unstimulated

#here you also see that the assumption is made dat the total expression does not change
#becasue we take it at 1 and the log is then 0
```

The MA plot is plotting the log 2 fold change (M) against the mean of normalized counts (A). The counts are normalized using the sample-specific size factors shown in the barplot above. 
These size factors were calculated as follows in DeSeq:

1. Calculate the mean of the counts per gene and across samples
2. Calculate the ratio of gene count in a sample and mean gene count across samples
3. The median of the ratio is the size factor for each sample
4. Divide all counts by the sample specific size factor to get the normalized counts

Sources:

https://hbctraining.github.io/DGE_workshop/lessons/02_DGE_count_normalization.html

```r
# identify genes with the desired adjusted p-value cut-off
DGEgenes <- rownames(subset(res.sorted, padj < 0.05))
length(DGEgenes)#again the almost 4500 genes that are significantly upregulated
head(DGEgenes, n=20)
```

```r
rlog.norm.counts<-rlog(all.counts.matrix)
rownames(rlog.norm.counts)<-rownames(all.counts.matrix)
head(rlog.norm.counts)
```

In DeSeq2 a shrinkage parameter was introduced to prevent that differences in low count genes were overestimated after applying the log transfomation, this is the rlog function.

```r
# extract rlog-transformed values into a matrix
hm.mat_DGEgenes <- rlog.norm.counts[DGEgenes,]
```

```r
library(pheatmap)
# heatmap of DEG sorted by p.adjust
not_scaled<-pheatmap(hm.mat_DGEgenes, scale="none", show_rownames = FALSE, main = "DGE (no scaling)")
scaled<-pheatmap(hm.mat_DGEgenes, scale="row", show_rownames = FALSE, main = "DGE (row-based z-score)")
```

The pheatmap package clusters the different samples (columns) together, as expected the treatment (NETs) from 1 cluster and the control group froms another cluster. The clustering for the rows is on gene level.
The values that are used for clustering are the rlog transformed values. Pheatmap calculates the mean rlog transformed value per row cluster and then calculates how many standard deviation each column lays below or above that cluster mean- And we see very clearly that the NETs and the normal rheumatic athritis cells cluster together.


```r
# add a column of NAs
res$diffexpressed <- "NA"
# if log2Foldchange > 1 and pvalue < 0.05, set as "UP" 
res$diffexpressed[res$log2FoldChange >  1 & res$pvalue < 0.05] <- "UP"
# if log2Foldchange < -1 and pvalue < 0.05, set as "DOWN"
res$diffexpressed[res$log2FoldChange < -1 & res$pvalue < 0.05] <- "DOWN"
```

```r
#create new column gene_symbol and delabel depending on FC and p value 
res$gene_symbol=rownames(res)
res$delabel <- NA
res$delabel[!is.na(res$log2FoldChange) 
            & (res$log2FoldChange>4| res$log2FoldChange< (-3))
            & !is.na(res$pvalue) 
            & -log10(res$pvalue)>(100)] <- res$gene_symbol[!is.na(res$log2FoldChange) 
                                               & (res$log2FoldChange>4| res$log2FoldChange< (-3))
                                               & !is.na(res$pvalue)
                                               & -log10(res$pvalue)>(100)]
```

```r
#prepare the data frame for plotting
df <- data.frame(log2FC = res$log2FoldChange,
                 neg.log10pvalue = -log10(res$pvalue),
                 diffexpressed=res$diffexpressed,
                 delabel=res$delabel)
#This is perhaps a bit unconventional but I did not know how to plot p-values of 0 and so I added the values 
#that had a pvalue of 0 at a log10 pvalue of 400 into the plot.
df$neg.log10pvalue[which(df$neg.log10pvalue=="Inf")]=rep(400,sum(df$neg.log10pvalue=="Inf", na.rm=T))
```

```r
library(ggplot2)
library(ggrepel)#for nice text labels

p<-ggplot(df, aes(x=log2FC, y=neg.log10pvalue, col=diffexpressed, label=delabel)) +
    geom_point(na.rm=T)+
    theme_classic()+
 geom_text_repel(na.rm=T) 

p2 <- p +   geom_vline(xintercept=c(-1, 1), col="red") +
            geom_hline(yintercept=-log10(0.05), col="red")+
            geom_hline(yintercept=375, col="red",linetype = 'dotted')


mycolors <- c("blue", "red", "black")
names(mycolors) <- c("DOWN", "UP", "NA")
p3 <- p2 + scale_colour_manual(values = mycolors)

# x axis limits
p4<-p3 + xlim(-15, 15)+ ylim(-10, 500)
p4
```

In the above plot the data points in red are upregulated genes with a p value below 0.05 an a fold change larger than 2. The labelled values have a a p value below 10^(-100) and are more than 2^4=16 fold upregulated. The solid red line interprete the cut-offs of 2 fold up/down regulation and a p-vaue of 0.05. 
The dashed line is perhaps a bit unconventional but there were a couple of genes that were differentially expressed with a p-value equalling 0 (probably underflow). When taking the logarithm this leads to a infinity value and the values are not plotted properly. The values with p-value 0 are plotted at a p-value equalling 10^(-400) arbritrily chosen.
